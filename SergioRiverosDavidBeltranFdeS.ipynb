# 2. Constante de Permitividad Elctrica.

La constante de permitividad elctrica es una propiedad fsica que describe el comportamiento de los materiales a ser sometidis a un campo elctrico, se denota con la letra griega  (Epsilon). Es primordial varias leyes como la de Coulomb y la ley de Gauss ya que son fundamentales para el estudio del electromagnetismo como interaccin entre cargas elctricas y campo elctrico.

En otros papeles es imporntate en fenmenos elctricos como la capacitancia donde determina la cantidad de carga elctrica que puede almacenarse en un capacitor en funcin de su rea, separacin y material dielctrico entre las placas.

En la ley de Coulomb su ecuacin:
$$F = \frac{K q_1q_2}{r^2}$$
Donde $F$ es la fuerza elctrica entre las dos cargas, $q_1q_2$ son las magnitudes de las cargas y $K$ es la constante de Coulomb representada con la ecuacin:
$$K = \frac{1}{4}$$
Podemos observar que aqu se requiere el uso de la constante de la permitividad elctrica.

En cuanto a la ley de Gauss establece que el flujo elctrico a travs de una superficie cerrada es proporcional a la carga neta encerrada en el volumen limitado por dicha superficie. La expresin matemtica de la ley de Gauss es:
$$_E = \frac{Q_{enc}}{}$$
Aqu podemos encontrar el uso de la constante ya que es necesaria en muchos casos.

En el vaco, la constante de permitividad elctrica tiene un valor aproximado de $\epsilon_08.854\cdot10^{12} F/m$. Esta constante se conoce como la permitividad del vaco. En diferentes medios en el vacio, la permitivad elctrica puede ser diferente y denotada por $_r$ donde $r$ es el ndice de refraccin del material ya que es una relacin del uno con el otro.

**Importante:**

Para observar la constante de permitividad elctrica en el cdigo se pide que se ingrese el valor de la permitividad relativa del material, si se quiere observar el valor de la constante en el vaco $_0$ se debe ingresar como valor de la permitividad relativa del material $1$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Permitividad elctrica en el vaco (F/m)
epsilon_0 = 8.854e-12

# Permitividad relativa de algunos materiales comunes
materiales = ['Vaco', 'Aire', 'Agua', 'Vidrio', 'Plstico', 'Mica']
permitividad_relativa = [1.0, 1.0006, 80.1, 4.9, 2.3, 5.4]

# Clculo de la constante de permitividad elctrica para cada material
epsilon = [epsilon_0 * er for er in permitividad_relativa]

# Crear la grfica
plt.figure(figsize=(10, 6))
plt.plot(permitividad_relativa, epsilon, marker='o', linestyle='-', color='g')

# Etiquetas y ttulo
plt.xlabel('Permitividad Relativa')
plt.ylabel('Constante de Permitividad Elctrica (F/m)')
plt.title('Constante de Permitividad Elctrica vs Permitividad Relativa')

# Anotar los puntos en la grfica
for i, txt in enumerate(materiales):
    plt.annotate(txt, (permitividad_relativa[i], epsilon[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Mostrar la grfica
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 3. Constante de Permeabilidad Magntica

La permeabilidad magntica es una constante fsica que describe la relacin entre el campo magntico y la densidad de corriente elctrica en el vaco. Esta constante es fundamental en la formulacin de las leyes del electromagnetismo, se denota como $_0$.

La constante tiene un valor de $_0410^{-7}Tm/A$ la cual surge de la ley de Ampere con la relacin entre el campo magnticoy la densidad de corriente $J$ en el vaco dada por:
$$B=_0J$$
Esta ley establece la relacin entre el campo magntico alrededor de un conductor con corriente y la densidad de corriente que fluye a travs de l. La forma integral de la ley de Ampre se expresa como $Bdl=_0I_{enc}$ donde $B$ s el campo magntico, $dl$ es un elemento infinitesimal de longitud en la trayectoria cerrada, $I_{enc}$ es la corriente encerrada por la trayectoria y $_0$ es la constante de permeabilidad magntica en el vaco.

La ley de Biot-Savart describe cmo se genera un campo magntico debido a una corriente elctrica en un conductor. La forma diferencial de la ley de Biot-Savart se expresa como:
$$dB=\frac{_0}{4}\frac{Idlr}{r^3}$$
donde $r$ es el vector de posicin desde el elemento de corriente hasta el punto donde se mide el campo magntico.

En el vaco, tiene un comportamiento lineal, homogneo y su permeabilidad se da por la siguiente manera:

# 4. La rapidez de las Ondas Electromagnticas

 La rapidez de las ondas electromagnticas se refiere a la velocidad a la que estas ondas se propagan a travs del espacio. Las ondas electromagnticas son un fenmeno fundamental en la fsica que se producen por la oscilacin simultnea de campos elctricos y magnticos perpendiculares entre s.

 La velocidad en el vaco se denota como $c$ y es una constante universal que es $c299.792.458m/s$. La rapidez de las ondas electromagnticas es constante e independiente de la frecuencia o longitud de onda. Esto significa que todas las ondas electromagnticas, ya sean ondas de radio, microondas, infrarrojos, luz visible, ultravioleta, rayos X o rayos gamma, se propagan a la misma velocidad en el vaco.

Esta propiedad de la velocidad constante de las ondas electromagnticas es uno de los principios fundamentales de la teora de la relatividad especial de Einstein y tiene importantes implicaciones en la fsica moderna y la tecnologa, incluyendo la teora de las comunicaciones inalmbricas, la ptica, la astronoma y muchas otras reas.

 Las ondas electromagnticas, que incluyen la luz visible, las ondas de radio, los rayos X, los rayos gamma, entre otros, se propagan a travs del espacio a una velocidad constante, que es la velocidad de la luz en el vaco.

En medios materiales como el aire, el agua o los materiales slidos, la velocidad de las ondas electromagnticas puede ser menor que en el vaco. Esto se debe a que los campos elctricos y magnticos de las ondas interactan con los tomos y las molculas del medio, lo que puede causar una ralentizacin de la velocidad de propagacin. La magnitud de esta ralentizacin depende de las propiedades del medio, como su densidad, composicin y estructura.

En cuanto a las diferentes frecuencias de las ondas electromagnticas, como las ondas de radio, la luz visible, los rayos X y los rayos gamma, todas se propagan a la misma velocidad en el vaco. Sin embargo, tienen diferentes longitudes de onda y frecuencias asociadas. Por ejemplo, las ondas de radio tienen longitudes de onda ms largas y frecuencias ms bajas que la luz visible, mientras que los rayos X y los rayos gamma tienen longitudes de onda mucho ms cortas y frecuencias ms altas.
<img src="https://upload.wikimedia.org/wikipedia/commons/b/b3/EM_Spectrum_Properties_es.svg">

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
from matplotlib.colors import Normalize

# Constante: velocidad de la luz en el vaco (m/s)
c = 299792458

# Diccionario de diferentes tipos de ondas electromagnticas y sus longitudes de onda en el vaco (en metros)
ondas_electromagneticas = {
    'Radio': (1e-1, 'red'),
    'Microondas': (1e-3, 'orange'),
    'Infrarrojo': (7e-7, 'yellow'),
    'Luz visible': (4e-7, 'green'),
    'Ultravioleta': (1e-8, 'blue'),
    'Rayos X': (1e-11, 'indigo'),
    'Rayos Gamma': (1e-13, 'violet')
}

# Calcular la rapidez de las ondas electromagnticas para cada tipo y longitud de onda
longitudes_de_onda = np.logspace(-14, -5, 1000)  # Rango de longitudes de onda
velocidades = c / longitudes_de_onda

# Crear una figura interactiva
fig, ax = plt.subplots(figsize=(12, 8))

# Graficar las ondas electromagnticas
for nombre, (longitud, color) in ondas_electromagneticas.items():
    ax.scatter(longitud, c / longitud, label=nombre, color=color, s=100, edgecolor='k', zorder=2)

# Etiquetas y leyenda
ax.set_title('Rapidez de las Ondas Electromagnticas en el Vaco')
ax.set_xlabel('Longitud de Onda (m)')
ax.set_ylabel('Rapidez (m/s)')
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid(True)
ax.legend(loc='upper right', fontsize=10)

# Aadir etiquetas a los puntos
for nombre, (longitud, _) in ondas_electromagneticas.items():
    ax.text(longitud, c / longitud, nombre, fontsize=8, va='bottom', ha='center')

# Aadir barra de color para representar las longitudes de onda
cmap = get_cmap('nipy_spectral')
norm = Normalize(vmin=np.min(longitudes_de_onda), vmax=np.max(longitudes_de_onda))
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, orientation='vertical', fraction=0.05, pad=0.05)
cbar.set_label('Longitud de Onda (m)', fontsize=10)

plt.tight_layout()
plt.show()
```

# 5. Carga del Electrn

La carga del electrn es una propiedad fundamental de las partculas subatmucas llamadas electrones. Los electrones son partculas elementales que orbitan alrededor del ncleo de un tomo y tienen una carga elctrica negativa. Esta carga es una de las dos propiedades bsicas de los electrones, siendo la otra su masa. Es fundamental en la fsica y se denota con la letra $e$. La carga de un electrn es de aproximadamente $e-1.60210^{-19}C$ (culombios) es importante destacar que la carga del electrn es una carga elemental, lo que significa que no puede dividirse en partes ms pequeas. La carga del electrn juega un papel crucial en muchos fenmenos elctricos y electromagnticos en la naturaleza, desde la interaccin entre tomos y molculas hasta la conduccin de electricidad a travs de materiales conductores.

La carga del electrn es usada en la ley de Coulomb ya que establece la unidad bsica de carga.  La magnitud de la fuerza entre las cargas es directamente proporcional al producto de las magnitudes de las cargas e inversamente proporcional al cuadrado de la distancia entre ellas.

Para la ley de Ohm se establece la relacin entre la corriente elctrica, la diferencia de potencial y la resistencia elctrica en un circuito. La carga del electrn se utiliza para comprender la corriente elctrica, que es el flujo de cargas elctricas (en este caso, electrones) a travs de un conductor.

Para la ley de Gauss establece que el flujo elctrico a travs de una superficie cerrada es proporcional a la carga neta encerrada dentro de la superficie. La carga del electrn se utiliza para definir la unidad bsica de carga en esta ley.

Y para la ley de Faraday en la induccin electromagntica se describe cmo un cambio en el flujo magntico a travs de un circuito induce un voltaje electromotriz en el circuito. La carga del electrn es fundamental en esta ley, ya que explica cmo la interaccin entre las cargas elctricas y los campos magnticos puede producir corriente elctrica.



```python
import matplotlib.pyplot as plt

# Definimos la carga del electrn en culombios
carga_electron = -1.602e-19  # en culombios

# Creamos un grfico de dispersin para representar la carga del electrn
plt.figure(figsize=(8, 6))
plt.scatter(0, 0, s=100, color='blue', label='Electrn')
plt.text(0.005, 0.005, f'Carga: {carga_electron} C', fontsize=8)
plt.xlabel('Posicin en el eje x')
plt.ylabel('Posicin en el eje y')
plt.title('Representacin de la Carga del Electrn')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

En la anterior grfica observamos un electrn con su respectiva carga, ahora observaremos diferentes electrones al rededor afectandose, en realidad su carga se mantiene, sin embargo, su campo es diferente como se observar en la siguiente grfica:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la carga del electrn en culombios
carga_electron = -1.602e-19  # en culombios

# Definimos las posiciones de los electrones
num_electrones = 5
posiciones_x = np.linspace(-3, 3, num_electrones)
posiciones_y = np.linspace(-1, 1, num_electrones)

# Creamos el grfico
plt.figure(figsize=(8, 6))

# Calculamos y representamos el campo elctrico generado por cada electrn y su carga
for i in range(num_electrones):
    # Posicin del electrn
    x_electron = posiciones_x[i]
    y_electron = posiciones_y[i]

    # Clculo del campo elctrico generado por el electrn en su posicin
    xx, yy = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
    r = np.sqrt((xx - x_electron) ** 2 + (yy - y_electron) ** 2)
    Ex = carga_electron * (xx - x_electron) / r**3
    Ey = carga_electron * (yy - y_electron) / r**3

    # Graficamos el campo elctrico generado por el electrn
    plt.quiver(xx, yy, Ex, Ey, scale=100, color='red', alpha=0.5)

    # Representamos el electrn como un punto y mostramos su carga
    plt.scatter(x_electron, y_electron, s=100, color='blue', label='Electrn')
    plt.text(x_electron + 0.05, y_electron + 0.05, f'Carga: {carga_electron} C', fontsize=10)

# Configuracin del grfico
plt.xlabel('Posicin en el eje x')
plt.ylabel('Posicin en el eje y')
plt.title('Representacin de Electrones y sus Campos Elctricos')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

# 6. Masa del Electrn

La masa del electrn es una propiedad de la fsica en especial de la teora cuntica siendo una de las constantes ms conocidas. El electrn es una particula subatmica con carga negativa y una masa extremadamente pequea. SU masa se puede representar en unidades de masa atmica o en Kilogramos.

El valor de la masa del electrn es aproximadamente $9.1093835610^{-31}Kg$ o en unidades atmicas $0.0005485799uma$.

Este dato es necesario en numerosas reas de la fisica y la ciencia en general.
1. Fisica de Particulas: La masa del electrn es esencial en la descripcin de las propiedades de otras partculas subatmicas y en la comprensin de las interacciones fundamentales en el mbito de la fsica de partculas. Ayuda a definir la escala de masas en el mundo subatmico y a entender cmo interactan las partculas elementales.
2. Qumica: En qumica, la masa del electrn es crucial para comprender la estructura y el comportamiento de los tomos y las molculas. La forma en que los electrones se distribuyen alrededor del ncleo atmico y cmo interactan con otros tomos determina las propiedades qumicas de los elementos y compuestos.
3. Electrnica: En la tecnologa, la masa del electrn es importante para disear dispositivos electrnicos, entender su comportamiento y predecir cmo funcionarn en diferentes condiciones. Por ejemplo, los electrones son responsables del flujo de corriente elctrica en los circuitos electrnicos, y su masa influye en la velocidad a la que pueden moverse a travs de los materiales conductores.

#7.  Grfica R-J, Planck, Limite de Wine. (Solucin Ecuacin Trascendente)


La Grfica R-J (Rayleigh-Jeans), la ley de Planck y el lmite de Wien son pilares fundamentales en el desarrollo temprano de la mecnica cuntica. La Grfica R-J, propuesta por Lord Rayleigh y Sir James Jeans, intent describir la distribucin espectral de la radiacin del cuerpo negro en base a la teora electromagntica clsica, pero fall en predecir de manera precisa la radiacin en altas frecuencias, lo que result en lo que se conoce como la catstrofe del ultravioleta. Para abordar este problema, Max Planck introdujo su ley de radiacin del cuerpo negro, una solucin a la ecuacin trascendental que describe la distribucin espectral de la radiacin, revolucionando as nuestra comprensin de la naturaleza de la radiacin electromagntica y sentando las bases para el nacimiento de la mecnica cuntica. El lmite de Wien, derivado de la ley de Planck, describe la relacin entre la longitud de onda de pico y la temperatura de un cuerpo negro, proporcionando una aproximacin vlida para longitudes de onda cortas. Estos avances marcaron un hito crucial en la historia de la fsica, preparando el terreno para el desarrollo posterior de la mecnica cuntica fundacional.

La funcion de longitud de onda de Rayleigh-Jeans esta definida por:

$$ \rho_{T}(\lambda)d\lambda=\frac{8\pi KT}{\lambda^{5}} \lambda d\lambda $$

Segun la teoria de planck

$$ \rho_{T}(\lambda)d\lambda=\frac{8\pi hc}{\lambda^{5}} \frac{1}{e^{\frac{hc}{\lambda KT}}-1} $$



```python
import math
from matplotlib.pyplot import figure
import numpy as np
import matplotlib.pyplot as plt

c1=3.7418*10**(-16) #
c2=1.4388*10**(-2) #
c3=8.28*10**(-15) #
c4=2897.6*10**(-6) #
T1=4000 #Temperatura 1
T2=5000 #Temperatura 2
T3=6000 #Temperatura 3
l=np.arange(100*10**(-9),3000*10**(-9),10**(-9))
T=np.arange(10,500000,10)
lmax=c4/T
Q1=c1/((pow(l,5))*(np.exp(c2/(l*T1))-1))
Q2=c1/((pow(l,5))*(np.exp(c2/(l*T2))-1))
Q3=c1/((pow(l,5))*(np.exp(c2/(l*T3))-1))
RJ1=c3*T1/((pow(l,4)))
RJ2=c3*T2/((pow(l,4)))
RJ3=c3*T3/((pow(l,4)))
Wien=c1*np.exp(-c2/(lmax*T))/(lmax**5)
plt.plot(l*10**9,Q1)
plt.plot(l*10**9,RJ1,"--")
plt.plot(l*10**9,Q2)
plt.plot(l*10**9,RJ2,"--")
plt.plot(l*10**9,Q3)
plt.plot(l*10**9,RJ3,"--")
plt.plot(lmax*10**9,Wien)

plt.legend(["Planck T=4000k","Rayleigh-Jeans T=4000k","Planck T=5000k","Rayleigh-Jeans T=5000k","Planck T=6000k","Rayleigh-Jeans T=6000k","Limite de Wien"])
plt.xlabel("$\lambda(nm)$")
plt.ylabel("$I_v(J/m^2)$")
plt.ylim((0,0.15*10**(15)))
plt.xlim((0,3000))
plt.title("Funcin de Planck, Aproximacin de Rayleigh-Jeans, Limite de Wien" )
plt.rcParams["figure.figsize"] = (10,6)


plt.show()
```

El lmite de Wien es un principio fsico que establece la relacin entre la temperatura de un objeto que emite radiacin trmica y la longitud de onda en la que se produce la mxima emisin de energa. Este principio fue descubierto por Wilhelm Wien en 1896 y es una consecuencia directa de la ley de desplazamiento de Wien.

El lmite de Wien establece que la longitud de onda de pico (max), es decir, la longitud de onda en la que se produce la mxima emisin de energa por parte de un objeto caliente, est inversamente relacionada con la temperatura absoluta (T) del objeto. Matemticamente, se expresa como:

$$1-e^{-x}-\frac{1}{5}x=0$$

Esta es uan ecuacion ecuacin trascendente, por tal motivo no tiene solucion analitica y sebe usar metodos numericos para solucionarla. Una forma de solucionarla es con ayuda de algoritmos.

```python
import math
import warnings
from scipy.optimize import fsolve

# Constantes
h = 6.62607015e-34  # Constante de Planck [m^2 kg / s]
c = 299792458       # Velocidad de la luz [m/s]
K = 1.380649e-23    # Constante de Boltzmann [m^2 kg / s^2 K]
T = 300             # Temperatura [K]

# Funcin para la ecuacin trascendental
def ecuacion(x):
    return -math.exp(-x) - (1/5)*x

# Usar fsolve con parmetros ajustados
solucion = fsolve(ecuacion, 5, xtol=1e-8, maxfev=1000)

# Extraer el nico elemento del array
solucion = solucion[0]

# Suprimir la advertencia de deprecacin
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    print("Solucin encontrada:", solucion)

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import newton
from matplotlib.widgets import Cursor

def f(x):
    return (x/5) - 1 + np.exp(-x)

def create_functions(x):
    f1 = 1 - np.exp(-x)
    f2 = x/5
    f3 = 1 - np.exp(-x) - x/5
    return f1, f2, f3


def update_graph(event):
    global x0, c, ax, tx, co
    if event.key in ['up', 'down', 'left', 'right']:
        if event.key == 'up':
            c = (c + 1) % 3
        elif event.key == 'down':
            c = (c - 1) % 3
        elif event.key == 'left':
            x0 -= 0.1
        elif event.key == 'right':
            x0 += 0.1

        f1, f2, f3 = create_functions(x)
        if c == 0:
            y0 = 1 - np.exp(-x0)
            tx = r'$1-e^{-x}$'
            co = 'red'
        elif c == 1:
            y0 = x0 / 5
            tx = r'$\frac{x}{5}$'
            co = 'purple'
        else:
            y0 = 1 - np.exp(-x0) - x0/5
            tx = r'$1-e^{-x}-\frac{x}{5}$'
            co = 'yellow'

        ax.lines[0].set_ydata(f1)
        ax.lines[1].set_ydata(f2)
        ax.lines[2].set_ydata(f3)

        ax.legend(labels=[r'$1-e^{-x}$', r'$\frac{x}{5}$', r'$1-e^{-x}-\frac{x}{5}$', tx])
        ax.axvline(x=4.96511, color='gray', linestyle='--')  # Lnea vertical en x=4.96511
        ax.text(4.96511, 0.5, 'x=4.96511', fontsize=8, color='gray')  # Marca en x=4.96511

        plt.draw()

x = np.arange(0, 7, 0.05)
s = newton(f, 5)
x0 = s
c = 2  # Initial equation chosen
tx = "Equation to choose"
co = "blue"

fig, ax = plt.subplots(figsize=(10, 6))
plt.grid()

f1, f2, f3 = create_functions(x)
plt.plot(x, f1, color='purple', linewidth=1.8, label=r'$1-e^{-x}$')
plt.plot(x, f2, color='black', linewidth=1.8, label=r'$\frac{x}{5}$')
plt.plot(x, f3, color='red', linewidth=1.8, label=r'$1-e^{-x}-\frac{x}{5}$')

cursor = Cursor(ax, horizOn=True, vertOn=True, color="black", linewidth=1.8)

ax.legend(labels=[r'$1-e^{-x}$', r'$\frac{x}{5}$', r'$1-e^{-x}-\frac{x}{5}$', tx])
plt.title("Funcin de Planck " + '$(\lambda)$')
plt.legend(loc='upper right', fontsize=11)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.text(0,1.2,'El punto de corte se encuentra en ' + r'x = $\frac{hc}{\lambda*kT}$ = %1.5f' %(s),fontsize = 12)
plt.axvline(x=4.96511, color='gray', linestyle='--')  # Lnea vertical en x=4.96511
plt.text(4.96511, 0, 'x=4.96511', fontsize=8, color='gray')  # Marca en x=4.96511
plt.connect('key_press_event', update_graph)
plt.show()
```

El efecto Compton proporcion evidencia experimental crucial para confirmar la naturaleza cuntica de la luz y el concepto de fotones, y ha sido fundamental en el desarrollo de teoras como la mecnica cuntica y la teora de la relatividad. #10. Modelo Atomico de Bohr
El modelo atmico de Niels Bohr, propuesto en 1913, revolucion nuestra comprensin de la estructura de los tomos y sent las bases para la mecnica cuntica moderna. Segn este modelo, los electrones en un tomo se mueven en rbitas circulares alrededor del ncleo central, similar a los planetas que orbitan alrededor del sol. Bohr postul que estos electrones solo pueden ocupar ciertos niveles de energa discretos, o "rbitas estacionarias", y que no pueden irradiar energa mientras permanecen en estas rbitas. Sin embargo, pueden saltar entre rbitas estacionarias, absorbiendo o emitiendo energa en forma de fotones cuando lo hacen. Este modelo explic con xito el espectro discreto de lneas de emisin de los tomos, como el hidrgeno, y sent las bases para la comprensin moderna de la estructura atmica y la mecnica cuntica. Aunque el modelo de Bohr fue posteriormente reemplazado por teoras ms completas, sigue siendo fundamental para la enseanza y comprensin de la fsica cuntica.

![alt text ](https://modelosatomicos.top/wp-content/uploads/2020/12/Hacer-una-Maqueta-del-Modelo-Atomico-de-Bohr.png)

```python
import numpy as np
import matplotlib.pyplot as plt

n= np.linspace(1, 10, 10)
x=np.linspace(900,32000,10000)
fig, ax = plt.subplots()
nl = 1
nb = 1/4
np = 1/9
nbr = 1/16
npf = 1/25
for i in n:
  y= -13.6*(1/(i**2))
  ax.hlines(y, xmin=x.min(), xmax=x.max())
  dl=(nl-(1/(i**2)))
  lal=(1/(10967757.6*dl))*10**10
  db=(nb-(1/(i**2)))
  lab=(1/(10967757.6*db))*10**10
  dp=(np-(1/(i**2)))
  lap=(1/(10967757.6*dp))*10**10
  dbr=(nbr-(1/(i**2)))
  labr=(1/(10967757.6*dbr))*10**10
  dpf=(npf-(1/(i**2)))
  lapf=(1/(10967757.6*dpf))*10**10
  if(dl>0):
    ax.vlines(lal, ymin= -13.6, ymax=y, colors='red')
  if(db>0):
    ax.vlines(lab, ymin= -3.4, ymax=y, colors='yellow')
  if(dp>0):
    ax.vlines(lap, ymin= -1.511, ymax=y, colors='green')
  if(dbr>0):
    ax.vlines(labr, ymin= -0.85, ymax=y, colors='blue')
  if(dpf>0):
    ax.vlines(lapf, ymin= -0.544, ymax=y, colors='violet')

ax.set_xscale('log')
plt.title('Series espectrales')
plt.ylabel('niveles de energa')
plt.xlabel('longitud de onda')
plt.axis(xmin=x.min()-100,xmax=x.max()+100, ymin=-14, ymax=0)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
R = 1.097e-2  # Constante de Rydberg en nm^-1

# Funcin para calcular las longitudes de onda
def calculate_wavelength(n1, n2):
    return 1 / (R * (1/n1**2 - 1/n2**2))

# Serie de Lyman: n1=1, n2=2,3,4,...
n_lyman = np.arange(2, 6)  # Calculando para los primeros 4 niveles
wavelength_lyman = calculate_wavelength(1, n_lyman)

# Serie de Balmer: n1=2, n2=3,4,5,...
n_balmer = np.arange(3, 7)  # Calculando para los primeros 4 niveles
wavelength_balmer = calculate_wavelength(2, n_balmer)

# Serie de Paschen: n1=3, n2=4,5,6,...
n_paschen = np.arange(4, 8)  # Calculando para los primeros 4 niveles
wavelength_paschen = calculate_wavelength(3, n_paschen)

# Graficando las series espectrales
plt.figure(figsize=(10, 6))
plt.plot(n_lyman, wavelength_lyman, 'bo-', label='Serie de Lyman')
plt.plot(n_balmer, wavelength_balmer, 'ro-', label='Serie de Balmer')
plt.plot(n_paschen, wavelength_paschen, 'go-', label='Serie de Paschen')
plt.xlabel('Nivel Final (n)')
plt.ylabel('Longitud de onda (nm)')
plt.title('Series espectrales del tomo de hidrgeno segn el modelo de Bohr')
plt.legend()
plt.grid(True)
plt.show()
```

#11. Experimento de franck-Hertz
El experimento de Franck-Hertz, realizado por James Franck y Gustav Hertz en 1914, fue un hito crucial en la confirmacin experimental de la teora cuntica. En este experimento, los cientficos bombardearon tomos de mercurio con electrones de baja energa y observaron la dispersin de estos electrones a medida que interactuaban con los tomos de mercurio. Descubrieron que los electrones solo perdan energa de manera significativa cuando su energa cintica era suficiente para excitar los tomos de mercurio a un nivel de energa especfico. Este fenmeno se manifestaba como una serie de picos en la corriente de electrones medida, donde cada pico corresponda a una cantidad especfica de energa necesaria para excitar los tomos de mercurio a un nivel superior. El experimento de Franck-Hertz proporcion evidencia directa de la cuantizacin de la energa en los tomos y valid la teora cuntica, confirmando as la idea de que la energa de los electrones y tomos est cuantizada en mltiplos enteros de una cantidad fundamental. Este experimento tuvo un impacto profundo en el desarrollo de la fsica cuntica y ayud a establecer los fundamentos de la mecnica cuntica.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del experimento
V_max = 50  # Voltaje mximo aplicado en voltios
V_step = 0.1  # Incremento de voltaje en voltios
V_range = np.arange(0, V_max + V_step, V_step)  # Rango de voltajes
I = np.zeros_like(V_range)  # Corriente medida

# Parmetros del tomo de mercurio
E_excitacion = 4.9  # Energa de excitacin del mercurio en electronvoltios (eV)
num_picos = int(V_max / E_excitacion)  # Nmero de picos esperados

# Simulacin del experimento
for i, V in enumerate(V_range):
    if V % E_excitacion < V_step:  # Solo excita cuando alcanza un mltiplo entero de la energa de excitacin
        I[i] = num_picos
    else:
        I[i] = 0

# Graficando los resultados
plt.plot(V_range, I)
plt.xlabel('Voltaje (V)')
plt.ylabel('Corriente medida')
plt.title('Simulacin del Experimento de Franck-Hertz')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del experimento
num_electrones = 10000  # Nmero de electrones simulados
energia_inicial = 10.0  # Energa inicial de los electrones en electronvoltios (eV)
energia_excitacion = 4.9  # Energa de excitacin del mercurio en electronvoltios (eV)

# Simulacin de la interaccin de los electrones con tomos de mercurio
energias_finales = np.zeros(num_electrones)

for i in range(num_electrones):
    energia = energia_inicial
    while energia > 0:
        energia -= np.random.exponential(energia_excitacion)  # Prdida de energa aleatoria
    energias_finales[i] = energia

# Graficando la distribucin de energa final de los electrones
plt.hist(energias_finales, bins=50, color='blue', edgecolor='black')
plt.xlabel('Energa Final (eV)')
plt.ylabel('Frecuencia')
plt.title('Distribucin de Energa Final de los Electrones')
plt.grid(True)
plt.show()
```
