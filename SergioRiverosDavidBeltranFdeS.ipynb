#1. Introduccion
Durante la primera mitad del siglo XX, la física experimentó una revolución sin precedentes que trastornó las bases mismas de nuestra comprensión del mundo subatómico. En un principio, la comunidad científica se regocijaba en la creencia de que los fundamentos de la física clásica eran suficientes para explicar todos los fenómenos naturales. Sin embargo, este optimismo se vio abruptamente interrumpido por la llamada 'catástrofe del ultravioleta', un fenómeno inesperado y desafiante que surgió en el estudio de la radiación del cuerpo negro. Los experimentos revelaron discrepancias significativas entre las predicciones teóricas basadas en las leyes clásicas de la física y los datos observados a frecuencias ultravioleta. Este desconcertante desajuste marcó el inicio de una nueva era en la física, catapultando a brillantes mentes como Max Planck, Albert Einstein y Niels Bohr hacia la formulación de teorías cuánticas revolucionarias que transformarían nuestra comprensión del comportamiento de la materia y la energía a niveles subatómicos.

# 2. Constante de Permitividad Eléctrica.

La constante de permitividad eléctrica es una propiedad física que describe el comportamiento de los materiales a ser sometidis a un campo eléctrico, se denota con la letra griega ϵ (Epsilon). Es primordial varias leyes como la de Coulomb y la ley de Gauss ya que son fundamentales para el estudio del electromagnetismo como interacción entre cargas eléctricas y campo eléctrico.

En otros papeles es imporntate en fenómenos eléctricos como la capacitancia donde determina la cantidad de carga eléctrica que puede almacenarse en un capacitor en función de su área, separación y material dieléctrico entre las placas.

En la ley de Coulomb su ecuación:
$$F = \frac{K⋅ q_1⋅q_2}{r^2}$$
Donde $F$ es la fuerza eléctrica entre las dos cargas, $q_1q_2$ son las magnitudes de las cargas y $K$ es la constante de Coulomb representada con la ecuación:
$$K = \frac{1}{4πϵ}$$
Podemos observar que aquí se requiere el uso de la constante de la permitividad eléctrica.

En cuanto a la ley de Gauss establece que el flujo eléctrico a través de una superficie cerrada es proporcional a la carga neta encerrada en el volumen limitado por dicha superficie. La expresión matemática de la ley de Gauss es:
$$Φ_E = \frac{Q_{enc}}{ϵ}$$
Aquí podemos encontrar el uso de la constante ya que es necesaria en muchos casos.

En el vacío, la constante de permitividad eléctrica tiene un valor aproximado de $\epsilon_0≈8.854\cdot10^{12} F/m$. Esta constante se conoce como la permitividad del vacío. En diferentes medios en el vacio, la permitivad eléctrica puede ser diferente y denotada por $ϵ_r$ donde $r$ es el índice de refracción del material ya que es una relación del uno con el otro.

**Importante:**

Para observar la constante de permitividad eléctrica en el código se pide que se ingrese el valor de la permitividad relativa del material, si se quiere observar el valor de la constante en el vacío $ϵ_0$ se debe ingresar como valor de la permitividad relativa del material $1$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Permitividad eléctrica en el vacío (F/m)
epsilon_0 = 8.854e-12

# Permitividad relativa de algunos materiales comunes
materiales = ['Vacío', 'Aire', 'Agua', 'Vidrio', 'Plástico', 'Mica']
permitividad_relativa = [1.0, 1.0006, 80.1, 4.9, 2.3, 5.4]

# Cálculo de la constante de permitividad eléctrica para cada material
epsilon = [epsilon_0 * er for er in permitividad_relativa]

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(permitividad_relativa, epsilon, marker='o', linestyle='-', color='g')

# Etiquetas y título
plt.xlabel('Permitividad Relativa')
plt.ylabel('Constante de Permitividad Eléctrica (F/m)')
plt.title('Constante de Permitividad Eléctrica vs Permitividad Relativa')

# Anotar los puntos en la gráfica
for i, txt in enumerate(materiales):
    plt.annotate(txt, (permitividad_relativa[i], epsilon[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Mostrar la gráfica
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 3. Constante de Permeabilidad Magnética

La permeabilidad magnética es una constante física que describe la relación entre el campo magnético y la densidad de corriente eléctrica en el vacío. Esta constante es fundamental en la formulación de las leyes del electromagnetismo, se denota como $μ_0$.

La constante tiene un valor de $μ_0≈4π⋅10^{-7}T⋅m/A$ la cual surge de la ley de Ampere con la relación entre el campo magnéticoy la densidad de corriente $J$ en el vacío dada por:
$$B=μ_0⋅J$$
Esta ley establece la relación entre el campo magnético alrededor de un conductor con corriente y la densidad de corriente que fluye a través de él. La forma integral de la ley de Ampère se expresa como $∮B⋅dl=μ_0⋅I_{enc}$ donde $B$ s el campo magnético, $dl$ es un elemento infinitesimal de longitud en la trayectoria cerrada, $I_{enc}$ es la corriente encerrada por la trayectoria y $μ_0$ es la constante de permeabilidad magnética en el vacío.

La ley de Biot-Savart describe cómo se genera un campo magnético debido a una corriente eléctrica en un conductor. La forma diferencial de la ley de Biot-Savart se expresa como:
$$dB=\frac{μ_0}{4π}\frac{Idl⋅r}{r^3}$$
donde $r$ es el vector de posición desde el elemento de corriente hasta el punto donde se mide el campo magnético.

En el vacío, tiene un comportamiento lineal, homogéneo y su permeabilidad se da por la siguiente manera:

# 4. La rapidez de las Ondas Electromagnéticas

 La rapidez de las ondas electromagnéticas se refiere a la velocidad a la que estas ondas se propagan a través del espacio. Las ondas electromagnéticas son un fenómeno fundamental en la física que se producen por la oscilación simultánea de campos eléctricos y magnéticos perpendiculares entre sí.

 La velocidad en el vacío se denota como $c$ y es una constante universal que es $c≈299.792.458m/s$. La rapidez de las ondas electromagnéticas es constante e independiente de la frecuencia o longitud de onda. Esto significa que todas las ondas electromagnéticas, ya sean ondas de radio, microondas, infrarrojos, luz visible, ultravioleta, rayos X o rayos gamma, se propagan a la misma velocidad en el vacío.

Esta propiedad de la velocidad constante de las ondas electromagnéticas es uno de los principios fundamentales de la teoría de la relatividad especial de Einstein y tiene importantes implicaciones en la física moderna y la tecnología, incluyendo la teoría de las comunicaciones inalámbricas, la óptica, la astronomía y muchas otras áreas.

 Las ondas electromagnéticas, que incluyen la luz visible, las ondas de radio, los rayos X, los rayos gamma, entre otros, se propagan a través del espacio a una velocidad constante, que es la velocidad de la luz en el vacío.

En medios materiales como el aire, el agua o los materiales sólidos, la velocidad de las ondas electromagnéticas puede ser menor que en el vacío. Esto se debe a que los campos eléctricos y magnéticos de las ondas interactúan con los átomos y las moléculas del medio, lo que puede causar una ralentización de la velocidad de propagación. La magnitud de esta ralentización depende de las propiedades del medio, como su densidad, composición y estructura.

En cuanto a las diferentes frecuencias de las ondas electromagnéticas, como las ondas de radio, la luz visible, los rayos X y los rayos gamma, todas se propagan a la misma velocidad en el vacío. Sin embargo, tienen diferentes longitudes de onda y frecuencias asociadas. Por ejemplo, las ondas de radio tienen longitudes de onda más largas y frecuencias más bajas que la luz visible, mientras que los rayos X y los rayos gamma tienen longitudes de onda mucho más cortas y frecuencias más altas.
<img src="https://upload.wikimedia.org/wikipedia/commons/b/b3/EM_Spectrum_Properties_es.svg">

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
from matplotlib.colors import Normalize

# Constante: velocidad de la luz en el vacío (m/s)
c = 299792458

# Diccionario de diferentes tipos de ondas electromagnéticas y sus longitudes de onda en el vacío (en metros)
ondas_electromagneticas = {
    'Radio': (1e-1, 'red'),
    'Microondas': (1e-3, 'orange'),
    'Infrarrojo': (7e-7, 'yellow'),
    'Luz visible': (4e-7, 'green'),
    'Ultravioleta': (1e-8, 'blue'),
    'Rayos X': (1e-11, 'indigo'),
    'Rayos Gamma': (1e-13, 'violet')
}

# Calcular la rapidez de las ondas electromagnéticas para cada tipo y longitud de onda
longitudes_de_onda = np.logspace(-14, -5, 1000)  # Rango de longitudes de onda
velocidades = c / longitudes_de_onda

# Crear una figura interactiva
fig, ax = plt.subplots(figsize=(12, 8))

# Graficar las ondas electromagnéticas
for nombre, (longitud, color) in ondas_electromagneticas.items():
    ax.scatter(longitud, c / longitud, label=nombre, color=color, s=100, edgecolor='k', zorder=2)

# Etiquetas y leyenda
ax.set_title('Rapidez de las Ondas Electromagnéticas en el Vacío')
ax.set_xlabel('Longitud de Onda (m)')
ax.set_ylabel('Rapidez (m/s)')
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid(True)
ax.legend(loc='upper right', fontsize=10)

# Añadir etiquetas a los puntos
for nombre, (longitud, _) in ondas_electromagneticas.items():
    ax.text(longitud, c / longitud, nombre, fontsize=8, va='bottom', ha='center')

# Añadir barra de color para representar las longitudes de onda
cmap = get_cmap('nipy_spectral')
norm = Normalize(vmin=np.min(longitudes_de_onda), vmax=np.max(longitudes_de_onda))
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, orientation='vertical', fraction=0.05, pad=0.05)
cbar.set_label('Longitud de Onda (m)', fontsize=10)

plt.tight_layout()
plt.show()
```

# 5. Carga del Electrón

La carga del electrón es una propiedad fundamental de las partículas subatómucas llamadas electrones. Los electrones son partículas elementales que orbitan alrededor del núcleo de un átomo y tienen una carga eléctrica negativa. Esta carga es una de las dos propiedades básicas de los electrones, siendo la otra su masa. Es fundamental en la física y se denota con la letra $e$. La carga de un electrón es de aproximadamente $e≈-1.602⋅10^{-19}C$ (culombios) es importante destacar que la carga del electrón es una carga elemental, lo que significa que no puede dividirse en partes más pequeñas. La carga del electrón juega un papel crucial en muchos fenómenos eléctricos y electromagnéticos en la naturaleza, desde la interacción entre átomos y moléculas hasta la conducción de electricidad a través de materiales conductores.

La carga del electrón es usada en la ley de Coulomb ya que establece la unidad básica de carga.  La magnitud de la fuerza entre las cargas es directamente proporcional al producto de las magnitudes de las cargas e inversamente proporcional al cuadrado de la distancia entre ellas.

Para la ley de Ohm se establece la relación entre la corriente eléctrica, la diferencia de potencial y la resistencia eléctrica en un circuito. La carga del electrón se utiliza para comprender la corriente eléctrica, que es el flujo de cargas eléctricas (en este caso, electrones) a través de un conductor.

Para la ley de Gauss establece que el flujo eléctrico a través de una superficie cerrada es proporcional a la carga neta encerrada dentro de la superficie. La carga del electrón se utiliza para definir la unidad básica de carga en esta ley.

Y para la ley de Faraday en la inducción electromagnética se describe cómo un cambio en el flujo magnético a través de un circuito induce un voltaje electromotriz en el circuito. La carga del electrón es fundamental en esta ley, ya que explica cómo la interacción entre las cargas eléctricas y los campos magnéticos puede producir corriente eléctrica.



```python
import matplotlib.pyplot as plt

# Definimos la carga del electrón en culombios
carga_electron = -1.602e-19  # en culombios

# Creamos un gráfico de dispersión para representar la carga del electrón
plt.figure(figsize=(8, 6))
plt.scatter(0, 0, s=100, color='blue', label='Electrón')
plt.text(0.005, 0.005, f'Carga: {carga_electron} C', fontsize=8)
plt.xlabel('Posición en el eje x')
plt.ylabel('Posición en el eje y')
plt.title('Representación de la Carga del Electrón')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

En la anterior gráfica observamos un electrón con su respectiva carga, ahora observaremos diferentes electrones al rededor afectandose, en realidad su carga se mantiene, sin embargo, su campo es diferente como se observará en la siguiente gráfica:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la carga del electrón en culombios
carga_electron = -1.602e-19  # en culombios

# Definimos las posiciones de los electrones
num_electrones = 5
posiciones_x = np.linspace(-3, 3, num_electrones)
posiciones_y = np.linspace(-1, 1, num_electrones)

# Creamos el gráfico
plt.figure(figsize=(8, 6))

# Calculamos y representamos el campo eléctrico generado por cada electrón y su carga
for i in range(num_electrones):
    # Posición del electrón
    x_electron = posiciones_x[i]
    y_electron = posiciones_y[i]

    # Cálculo del campo eléctrico generado por el electrón en su posición
    xx, yy = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
    r = np.sqrt((xx - x_electron) ** 2 + (yy - y_electron) ** 2)
    Ex = carga_electron * (xx - x_electron) / r**3
    Ey = carga_electron * (yy - y_electron) / r**3

    # Graficamos el campo eléctrico generado por el electrón
    plt.quiver(xx, yy, Ex, Ey, scale=100, color='red', alpha=0.5)

    # Representamos el electrón como un punto y mostramos su carga
    plt.scatter(x_electron, y_electron, s=100, color='blue', label='Electrón')
    plt.text(x_electron + 0.05, y_electron + 0.05, f'Carga: {carga_electron} C', fontsize=10)

# Configuración del gráfico
plt.xlabel('Posición en el eje x')
plt.ylabel('Posición en el eje y')
plt.title('Representación de Electrones y sus Campos Eléctricos')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

# 6. Masa del Electrón

La masa del electrón es una propiedad de la física en especial de la teoría cuántica siendo una de las constantes más conocidas. El electrón es una particula subatómica con carga negativa y una masa extremadamente pequeña. SU masa se puede representar en unidades de masa atómica o en Kilogramos.

El valor de la masa del electrón es aproximadamente $9.10938356⋅10^{-31}Kg$ o en unidades atómicas $0.0005485799uma$.

Este dato es necesario en numerosas áreas de la fisica y la ciencia en general.
1. Fisica de Particulas: La masa del electrón es esencial en la descripción de las propiedades de otras partículas subatómicas y en la comprensión de las interacciones fundamentales en el ámbito de la física de partículas. Ayuda a definir la escala de masas en el mundo subatómico y a entender cómo interactúan las partículas elementales.
2. Química: En química, la masa del electrón es crucial para comprender la estructura y el comportamiento de los átomos y las moléculas. La forma en que los electrones se distribuyen alrededor del núcleo atómico y cómo interactúan con otros átomos determina las propiedades químicas de los elementos y compuestos.

#7.  Gráfica R-J, Planck, Limite de Wine. (Solución Ecuación Trascendente)


La Gráfica R-J (Rayleigh-Jeans), la ley de Planck y el límite de Wien son pilares fundamentales en el desarrollo temprano de la mecánica cuántica. La Gráfica R-J, propuesta por Lord Rayleigh y Sir James Jeans, intentó describir la distribución espectral de la radiación del cuerpo negro en base a la teoría electromagnética clásica, pero falló en predecir de manera precisa la radiación en altas frecuencias, lo que resultó en lo que se conoce como la catástrofe del ultravioleta. Para abordar este problema, Max Planck introdujo su ley de radiación del cuerpo negro, una solución a la ecuación trascendental que describe la distribución espectral de la radiación, revolucionando así nuestra comprensión de la naturaleza de la radiación electromagnética y sentando las bases para el nacimiento de la mecánica cuántica. El límite de Wien, derivado de la ley de Planck, describe la relación entre la longitud de onda de pico y la temperatura de un cuerpo negro, proporcionando una aproximación válida para longitudes de onda cortas. Estos avances marcaron un hito crucial en la historia de la física, preparando el terreno para el desarrollo posterior de la mecánica cuántica fundacional.

La funcion de longitud de onda de Rayleigh-Jeans esta definida por:

$$ \rho_{T}(\lambda)d\lambda=\frac{8\pi KT}{\lambda^{5}} \lambda d\lambda $$

Segun la teoria de planck

$$ \rho_{T}(\lambda)d\lambda=\frac{8\pi hc}{\lambda^{5}} \frac{1}{e^{\frac{hc}{\lambda KT}}-1} $$



```python
import math
from matplotlib.pyplot import figure
import numpy as np
import matplotlib.pyplot as plt

c1=3.7418*10**(-16) #
c2=1.4388*10**(-2) #
c3=8.28*10**(-15) #
c4=2897.6*10**(-6) #
T1=4000 #Temperatura 1
T2=5000 #Temperatura 2
T3=6000 #Temperatura 3
l=np.arange(100*10**(-9),3000*10**(-9),10**(-9))
T=np.arange(10,500000,10)
lmax=c4/T
Q1=c1/((pow(l,5))*(np.exp(c2/(l*T1))-1))
Q2=c1/((pow(l,5))*(np.exp(c2/(l*T2))-1))
Q3=c1/((pow(l,5))*(np.exp(c2/(l*T3))-1))
RJ1=c3*T1/((pow(l,4)))
RJ2=c3*T2/((pow(l,4)))
RJ3=c3*T3/((pow(l,4)))
Wien=c1*np.exp(-c2/(lmax*T))/(lmax**5)
plt.plot(l*10**9,Q1)
plt.plot(l*10**9,RJ1,"--")
plt.plot(l*10**9,Q2)
plt.plot(l*10**9,RJ2,"--")
plt.plot(l*10**9,Q3)
plt.plot(l*10**9,RJ3,"--")
plt.plot(lmax*10**9,Wien)

plt.legend(["Planck T=4000k","Rayleigh-Jeans T=4000k","Planck T=5000k","Rayleigh-Jeans T=5000k","Planck T=6000k","Rayleigh-Jeans T=6000k","Limite de Wien"])
plt.xlabel("$\lambda(nm)$")
plt.ylabel("$I_v(J/m^2)$")
plt.ylim((0,0.15*10**(15)))
plt.xlim((0,3000))
plt.title("Función de Planck, Aproximación de Rayleigh-Jeans, Limite de Wien" )
plt.rcParams["figure.figsize"] = (10,6)


plt.show()
```

El límite de Wien es un principio físico que establece la relación entre la temperatura de un objeto que emite radiación térmica y la longitud de onda en la que se produce la máxima emisión de energía. Este principio fue descubierto por Wilhelm Wien en 1896 y es una consecuencia directa de la ley de desplazamiento de Wien.

El límite de Wien establece que la longitud de onda de pico (λmax), es decir, la longitud de onda en la que se produce la máxima emisión de energía por parte de un objeto caliente, está inversamente relacionada con la temperatura absoluta (T) del objeto. Matemáticamente, se expresa como:

$$1-e^{-x}-\frac{1}{5}x=0$$

Esta es uan ecuacion ecuación trascendente, por tal motivo no tiene solucion analitica y sebe usar metodos numericos para solucionarla. Una forma de solucionarla es con ayuda de algoritmos.

```python
import math
import warnings
from scipy.optimize import fsolve

# Constantes
h = 6.62607015e-34  # Constante de Planck [m^2 kg / s]
c = 299792458       # Velocidad de la luz [m/s]
K = 1.380649e-23    # Constante de Boltzmann [m^2 kg / s^2 K]
T = 300             # Temperatura [K]

# Función para la ecuación trascendental
def ecuacion(x):
    return -math.exp(-x) - (1/5)*x

# Usar fsolve con parámetros ajustados
solucion = fsolve(ecuacion, 5, xtol=1e-8, maxfev=1000)

# Extraer el único elemento del array
solucion = solucion[0]

# Suprimir la advertencia de deprecación
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    print("Solución encontrada:", solucion)

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import newton
from matplotlib.widgets import Cursor

def f(x):
    return (x/5) - 1 + np.exp(-x)

def create_functions(x):
    f1 = 1 - np.exp(-x)
    f2 = x/5
    f3 = 1 - np.exp(-x) - x/5
    return f1, f2, f3


def update_graph(event):
    global x0, c, ax, tx, co
    if event.key in ['up', 'down', 'left', 'right']:
        if event.key == 'up':
            c = (c + 1) % 3
        elif event.key == 'down':
            c = (c - 1) % 3
        elif event.key == 'left':
            x0 -= 0.1
        elif event.key == 'right':
            x0 += 0.1

        f1, f2, f3 = create_functions(x)
        if c == 0:
            y0 = 1 - np.exp(-x0)
            tx = r'$1-e^{-x}$'
            co = 'red'
        elif c == 1:
            y0 = x0 / 5
            tx = r'$\frac{x}{5}$'
            co = 'purple'
        else:
            y0 = 1 - np.exp(-x0) - x0/5
            tx = r'$1-e^{-x}-\frac{x}{5}$'
            co = 'yellow'

        ax.lines[0].set_ydata(f1)
        ax.lines[1].set_ydata(f2)
        ax.lines[2].set_ydata(f3)

        ax.legend(labels=[r'$1-e^{-x}$', r'$\frac{x}{5}$', r'$1-e^{-x}-\frac{x}{5}$', tx])
        ax.axvline(x=4.96511, color='gray', linestyle='--')  # Línea vertical en x=4.96511
        ax.text(4.96511, 0.5, 'x=4.96511', fontsize=8, color='gray')  # Marca en x=4.96511

        plt.draw()

x = np.arange(0, 7, 0.05)
s = newton(f, 5)
x0 = s
c = 2  # Initial equation chosen
tx = "Equation to choose"
co = "blue"

fig, ax = plt.subplots(figsize=(10, 6))
plt.grid()

f1, f2, f3 = create_functions(x)
plt.plot(x, f1, color='purple', linewidth=1.8, label=r'$1-e^{-x}$')
plt.plot(x, f2, color='black', linewidth=1.8, label=r'$\frac{x}{5}$')
plt.plot(x, f3, color='red', linewidth=1.8, label=r'$1-e^{-x}-\frac{x}{5}$')

cursor = Cursor(ax, horizOn=True, vertOn=True, color="black", linewidth=1.8)

ax.legend(labels=[r'$1-e^{-x}$', r'$\frac{x}{5}$', r'$1-e^{-x}-\frac{x}{5}$', tx])
plt.title("Función de Planck " + '$(\lambda)$')
plt.legend(loc='upper right', fontsize=11)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.text(0,1.2,'El punto de corte se encuentra en ' + r'x = $\frac{hc}{\lambda*kT}$ = %1.5f' %(s),fontsize = 12)
plt.axvline(x=4.96511, color='gray', linestyle='--')  # Línea vertical en x=4.96511
plt.text(4.96511, 0, 'x=4.96511', fontsize=8, color='gray')  # Marca en x=4.96511
plt.connect('key_press_event', update_graph)
plt.show()
```

$$e^{\frac{hc}{λkT}}-1=\frac{1}{5}\frac{hc}{λKT}e^{\frac{hc}{λkT}}$$
$$e^x-1=\frac{1}{5}xe^x$$

Este efecto fue observado y estudiado por primera vez por Heinrich Hertz en 1887 y posteriormente explicado teóricamente por Albert Einstein en 1905, lo que le valió el Premio Nobel de Física en 1921. Este fenómeno es fundamental en la comprensión de la naturaleza cuántica de la luz y ha tenido importantes aplicaciones en áreas como la fotocatálisis, la generación de energía solar y la tecnología de detección de luz.

#9. El efecto coptom
El efecto Compton, descubierto por el físico Arthur Compton en 1923, es un fenómeno fundamental en la interacción entre la radiación electromagnética y la materia. En este fenómeno, un fotón incidente choca con un electrón libre, transfiriéndole parte de su energía y cambiando su dirección de propagación. Como resultado de esta colisión, el fotón dispersado experimenta un aumento en su longitud de onda, lo que indica una disminución en su energía. Esta desviación en la longitud de onda del fotón dispersado es conocida como desplazamiento de Compton y es independiente del material con el que interactúe. El efecto Compton proporcionó evidencia experimental crucial para confirmar la naturaleza cuántica de la luz y el concepto de fotones, y ha sido fundamental en el desarrollo de teorías como la mecánica cuántica y la teoría de la relatividad. Además, este fenómeno tiene aplicaciones importantes en la medicina, la investigación de materiales y la física de partículas.

#10. Modelo Atomico de Bohr
El modelo atómico de Niels Bohr, propuesto en 1913, revolucionó nuestra comprensión de la estructura de los átomos y sentó las bases para la mecánica cuántica moderna. Según este modelo, los electrones en un átomo se mueven en órbitas circulares alrededor del núcleo central, similar a los planetas que orbitan alrededor del sol. Bohr postuló que estos electrones solo pueden ocupar ciertos niveles de energía discretos, o "órbitas estacionarias", y que no pueden irradiar energía mientras permanecen en estas órbitas. Sin embargo, pueden saltar entre órbitas estacionarias, absorbiendo o emitiendo energía en forma de fotones cuando lo hacen. Este modelo explicó con éxito el espectro discreto de líneas de emisión de los átomos, como el hidrógeno, y sentó las bases para la comprensión moderna de la estructura atómica y la mecánica cuántica. Aunque el modelo de Bohr fue posteriormente reemplazado por teorías más completas, sigue siendo fundamental para la enseñanza y comprensión de la física cuántica.

![alt text ](https://modelosatomicos.top/wp-content/uploads/2020/12/Hacer-una-Maqueta-del-Modelo-Atomico-de-Bohr.png)

```python
import numpy as np
import matplotlib.pyplot as plt

n= np.linspace(1, 10, 10)
x=np.linspace(900,32000,10000)
fig, ax = plt.subplots()
nl = 1
nb = 1/4
np = 1/9
nbr = 1/16
npf = 1/25
for i in n:
  y= -13.6*(1/(i**2))
  ax.hlines(y, xmin=x.min(), xmax=x.max())
  dl=(nl-(1/(i**2)))
  lal=(1/(10967757.6*dl))*10**10
  db=(nb-(1/(i**2)))
  lab=(1/(10967757.6*db))*10**10
  dp=(np-(1/(i**2)))
  lap=(1/(10967757.6*dp))*10**10
  dbr=(nbr-(1/(i**2)))
  labr=(1/(10967757.6*dbr))*10**10
  dpf=(npf-(1/(i**2)))
  lapf=(1/(10967757.6*dpf))*10**10
  if(dl>0):
    ax.vlines(lal, ymin= -13.6, ymax=y, colors='red')
  if(db>0):
    ax.vlines(lab, ymin= -3.4, ymax=y, colors='yellow')
  if(dp>0):
    ax.vlines(lap, ymin= -1.511, ymax=y, colors='green')
  if(dbr>0):
    ax.vlines(labr, ymin= -0.85, ymax=y, colors='blue')
  if(dpf>0):
    ax.vlines(lapf, ymin= -0.544, ymax=y, colors='violet')

ax.set_xscale('log')
plt.title('Series espectrales')
plt.ylabel('niveles de energía')
plt.xlabel('longitud de onda')
plt.axis(xmin=x.min()-100,xmax=x.max()+100, ymin=-14, ymax=0)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
R = 1.097e-2  # Constante de Rydberg en nm^-1

# Función para calcular las longitudes de onda
def calculate_wavelength(n1, n2):
    return 1 / (R * (1/n1**2 - 1/n2**2))

# Serie de Lyman: n1=1, n2=2,3,4,...
n_lyman = np.arange(2, 6)  # Calculando para los primeros 4 niveles
wavelength_lyman = calculate_wavelength(1, n_lyman)

# Serie de Balmer: n1=2, n2=3,4,5,...
n_balmer = np.arange(3, 7)  # Calculando para los primeros 4 niveles
wavelength_balmer = calculate_wavelength(2, n_balmer)

# Serie de Paschen: n1=3, n2=4,5,6,...
n_paschen = np.arange(4, 8)  # Calculando para los primeros 4 niveles
wavelength_paschen = calculate_wavelength(3, n_paschen)

# Graficando las series espectrales
plt.figure(figsize=(10, 6))
plt.plot(n_lyman, wavelength_lyman, 'bo-', label='Serie de Lyman')
plt.plot(n_balmer, wavelength_balmer, 'ro-', label='Serie de Balmer')
plt.plot(n_paschen, wavelength_paschen, 'go-', label='Serie de Paschen')
plt.xlabel('Nivel Final (n)')
plt.ylabel('Longitud de onda (nm)')
plt.title('Series espectrales del átomo de hidrógeno según el modelo de Bohr')
plt.legend()
plt.grid(True)
plt.show()
```

#11. Experimento de franck-Hertz
El experimento de Franck-Hertz, realizado por James Franck y Gustav Hertz en 1914, fue un hito crucial en la confirmación experimental de la teoría cuántica. En este experimento, los científicos bombardearon átomos de mercurio con electrones de baja energía y observaron la dispersión de estos electrones a medida que interactuaban con los átomos de mercurio. Descubrieron que los electrones solo perdían energía de manera significativa cuando su energía cinética era suficiente para excitar los átomos de mercurio a un nivel de energía específico. Este fenómeno se manifestaba como una serie de picos en la corriente de electrones medida, donde cada pico correspondía a una cantidad específica de energía necesaria para excitar los átomos de mercurio a un nivel superior. El experimento de Franck-Hertz proporcionó evidencia directa de la cuantización de la energía en los átomos y validó la teoría cuántica, confirmando así la idea de que la energía de los electrones y átomos está cuantizada en múltiplos enteros de una cantidad fundamental. Este experimento tuvo un impacto profundo en el desarrollo de la física cuántica y ayudó a establecer los fundamentos de la mecánica cuántica.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del experimento
V_max = 50  # Voltaje máximo aplicado en voltios
V_step = 0.1  # Incremento de voltaje en voltios
V_range = np.arange(0, V_max + V_step, V_step)  # Rango de voltajes
I = np.zeros_like(V_range)  # Corriente medida

# Parámetros del átomo de mercurio
E_excitacion = 4.9  # Energía de excitación del mercurio en electronvoltios (eV)
num_picos = int(V_max / E_excitacion)  # Número de picos esperados

# Simulación del experimento
for i, V in enumerate(V_range):
    if V % E_excitacion < V_step:  # Solo excita cuando alcanza un múltiplo entero de la energía de excitación
        I[i] = num_picos
    else:
        I[i] = 0

# Graficando los resultados
plt.plot(V_range, I)
plt.xlabel('Voltaje (V)')
plt.ylabel('Corriente medida')
plt.title('Simulación del Experimento de Franck-Hertz')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del experimento
num_electrones = 10000  # Número de electrones simulados
energia_inicial = 10.0  # Energía inicial de los electrones en electronvoltios (eV)
energia_excitacion = 4.9  # Energía de excitación del mercurio en electronvoltios (eV)

# Simulación de la interacción de los electrones con átomos de mercurio
energias_finales = np.zeros(num_electrones)

for i in range(num_electrones):
    energia = energia_inicial
    while energia > 0:
        energia -= np.random.exponential(energia_excitacion)  # Pérdida de energía aleatoria
    energias_finales[i] = energia

# Graficando la distribución de energía final de los electrones
plt.hist(energias_finales, bins=50, color='blue', edgecolor='black')
plt.xlabel('Energía Final (eV)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Energía Final de los Electrones')
plt.grid(True)
plt.show()
```
